```python
#import libraries

import matplotlib.patches as patches
import matplotlib.pyplot as plt
import numpy as np
```

```python
#defining Contrast Transfer Function (CTF) function

def calcCTF (*regions):
    CTFs = []
    for region in regions:
        CTF = (np.amax(region) - np.amin(region)) / (np.amax(region) + np.amin(region))
        CTFs.append(CTF)
    return CTFs;
```
