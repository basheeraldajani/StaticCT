```python
#import libraries

import matplotlib.patches as patches
import matplotlib.pyplot as plt
import numpy as np
from math import pi
```

```python
#defining Modulation Transfer Function (MTF) function

widths = [2, 2.5, 3.2, 4, 5, 6.3, 8, 10, 12.5, 16, 20] #LP/mm

def calcMTF (detector_layout, *widths):
    MTFs = []
    
    for width in widths:
        
        PCD_Vertical_a = 0.99536521590438
        PCD_Horizontal_a = 0.30703131755158
        Varian_Horizontal_a = 0.31727437497099
        
        if detector_layout == "PCD_Vertical":
            
            a = PCD_Vertical_a*0.92654688738608**width
            b = PCD_Vertical_a*0.92654688738608**(width*3)
            c = PCD_Vertical_a*0.92654688738608**(width*5)
            d = PCD_Vertical_a*0.92654688738608**(width*7)
            
        if detector_layout == "PCD_Horizontal":
            
            a = PCD_Horizontal_a*0.98169678739644**width
            b = PCD_Horizontal_a*0.98169678739644**(width*3)
            c = PCD_Horizontal_a*0.98169678739644**(width*5)
            d = PCD_Horizontal_a*0.98169678739644**(width*7)
        
        if detector_layout == "Varian_Horizontal":
            
            a = Varian_Horizontal_a*0.93087801631785**width
            b = Varian_Horizontal_a*0.93087801631785**(width*3)
            c = Varian_Horizontal_a*0.93087801631785**(width*5)
            d = Varian_Horizontal_a*0.93087801631785**(width*7)
            
        MTF = (pi/4) * ((a) + (b/3) - (c/5) + (d/7))
        MTFs.append(MTF)
        
    return MTFs;
```

```python
MTF1 = calcMTF("PCD_Vertical", widths[0], widths[1], widths[2], widths[3], widths[4], widths[5], widths[6], widths[7], widths[8], widths[9], widths[10])
print(MTF1)
```

```python
MTF2 = calcMTF("PCD_Horizontal", widths[0], widths[1], widths[2], widths[3], widths[4], widths[5], widths[6], widths[7], widths[8], widths[9], widths[10])
print(MTF2)
```

```python
MTF3 = calcMTF("Varian_Horizontal", widths[0], widths[1], widths[2], widths[3], widths[4], widths[5], widths[6], widths[7], widths[8], widths[9], widths[10])
print(MTF3)
```

```python
plt.plot(widths, MTF1, marker = 'o', markerfacecolor = 'blue', markersize = 4, color = "skyblue", linewidth = 2, label = 'PCD (Vertical)')
plt.plot(widths, MTF2, marker = 'o', markerfacecolor = 'red', markersize = 4, color = "pink", linewidth = 2, label = 'PCD (Horizontal)')
plt.plot(widths, MTF3, marker = 'o', markerfacecolor = 'green', markersize = 4, color = "olive", linewidth = 2, label = 'Varian (Horizontal)')
plt.title("Spatial Resolution (MTF)")
plt.xlabel("Spatial Frequency (LP/mm)")
plt.ylabel("Modulation Transfer Function (MTF)")
plt.legend(loc="best")
plt.show()
```
